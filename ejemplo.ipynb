**Cargar imagen original y visualizar con matplotlib**

```
# Lee y trae la imagen a color, no hace cambio en el alfa
img_BGR = cv2.imread('/content/drive/My Drive/13 - Trece semestre/Procesamiento de imagen/Prueba/Husky.jpg', cv2.IMREAD_COLOR)

plt.figure(figsize=(7,7)) # El 7,7 hace el tamaño del frame, el recuadro donde va la imagen
plt.imshow(img_BGR) # Lee la imagen modificada para mostrarla en el plot
plt.show() # Muestra la imagen en la pantalla
```

**Cambiar tamaño de la imagen**

```
# Expand (Expandir)
print(img.shape) # Imprime el tamaño de la imagen
res = cv2.resize(img, None, fx=10, fy=10, interpolation = cv2.INTER_CUBIC) # fx y fy me están dando el tamaño de la escala a la que quiero reescalar
print(res.shape) # Imprime el tamaño de la imagen modificada
plt.imshow(res) # Plotea el resultado 
plt.show()
```

```
# Shrinking (Contraer)

print(img.shape)
res = cv2.resize(img, None, fx=0.1, fy=0.1, interpolation = cv2.INTER_CUBIC) # fx y fy me permiten contraer la imagen al estar en 0.1
print(res.shape) # Imprime el tamaño de la imagen modificada
plt.imshow(res) # Plotea el resultado 
plt.show()
```

**Remover numeración de los ejes**

```
plt.figure(figsize=(7,7))
plt.imshow(img_BGR)
plt.xticks([])
plt.yticks([])
plt.show()
```

**Cambiar de BGR a RGB**

```
# Pasar la imagen a rgb
img_RGB=cv2.cvtColor(img_BGR,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.imshow(img_RGB)
plt.xticks([])
plt.yticks([])
plt.show()
```
