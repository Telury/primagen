**Cargar imagen original y visualizar con matplotlib**

```
# Lee y trae la imagen a color, no hace cambio en el alfa
img_BGR = cv2.imread('/content/drive/My Drive/13 - Trece semestre/Procesamiento de imagen/Prueba/Husky.jpg', cv2.IMREAD_COLOR)

plt.figure(figsize=(7,7)) # El 7,7 hace el tamaño del frame, el recuadro donde va la imagen
plt.imshow(img_BGR) # Lee la imagen modificada para mostrarla en el plot
plt.show() # Muestra la imagen en la pantalla
```

**Cambiar tamaño de la imagen**

```
# Expand (Expandir)
print(img.shape) # Imprime el tamaño de la imagen
res = cv2.resize(img, None, fx=10, fy=10, interpolation = cv2.INTER_CUBIC) # fx y fy me están dando el tamaño de la escala a la que quiero reescalar
print(res.shape) # Imprime el tamaño de la imagen modificada
plt.imshow(res) # Plotea el resultado 
plt.show()
```

```
# Shrinking (Contraer)

print(img.shape)
res = cv2.resize(img, None, fx=0.1, fy=0.1, interpolation = cv2.INTER_CUBIC) # fx y fy me permiten contraer la imagen al estar en 0.1
print(res.shape) # Imprime el tamaño de la imagen modificada
plt.imshow(res) # Plotea el resultado 
plt.show()
```

**Remover numeración de los ejes**

```
plt.figure(figsize=(7,7))
plt.imshow(img_BGR)
plt.xticks([])
plt.yticks([])
plt.show()
```

**Cambiar de BGR a RGB**

```
# Pasar la imagen a rgb
img_RGB=cv2.cvtColor(img_BGR,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.imshow(img_RGB)
plt.xticks([])
plt.yticks([])
plt.show()
```

**Convertir a escala de grises (Primer paso)**

```
# Pasar la imagen a gris
img_GRAY = cv2.cvtColor(img_BGR,cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10,10))
plt.imshow(img_GRAY)
plt.xticks([])
plt.yticks([])
plt.show()
```

**Convertir a escala de grises**

```
# Pasar la imagen a gris
plt.figure(figsize=(10,10))
plt.imshow(img_GRAY,cmap='gray')

plt.xticks([])
plt.yticks([])
plt.show()
```

**Usar métodos para mostrar las imágenes**

```
# Muestra dos imágenes
def display(a, b, title1 = "Original", title2 = "Edited"):
  plt.figure(figsize=(14,14))
  plt.subplot(121), plt.imshow(a), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.subplot(122), plt.imshow(b), plt.title(title2)
  plt.xticks([]), plt.yticks([])
  plt.show()
  
# Muestra una imagen
def display_one(a, title1 = "Original"):
  plt.imshow(a),plt.title(title1)
  plt.show()
  
display_one(img_BGR)  # Muestra una sola imagen
display(img_BGR, img_BGR) # Muestra dos imágenes
```

**Muestra cuatro imágenes**

```
# Muestra cuatro imágenes
def display4(a, b, c, d, title1 = "Original", title2 = "Edited", title3 = "Edited 3", title4 = "Edited 4"):
  plt.figure(figsize=(14,14))
  plt.subplot(221), plt.imshow(a), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.subplot(222), plt.imshow(b), plt.title(title2)
  plt.xticks([]), plt.yticks([])
  plt.subplot(223), plt.imshow(c), plt.title(title3)
  plt.xticks([]), plt.yticks([])
  plt.subplot(224), plt.imshow(d, cmap='gray'), plt.title(title4)
  plt.xticks([]), plt.yticks([])
  plt.show()
  
display4(img_BGR, img_RGB, img_GRAY, img_GRAY, "BRG", "RGB", "GRAY", "GRAY") # Muestra dos imágenes
```

**Características de la imagen**

```
def estadisticas(img):
  img_size=img.shape
  
  if len(img_size)>2:
    print("image chanels", img_size[2])
  else:
    print("image chanels",1)
  
  print("image W", img_size[1])
  print("image H", img_size[1])
  print("min value", img.min())
  print("max value", img.max())
  print("prom", img.mean())
  print("type", img.dtype)
  
estadisticas(img_BGR)   # Envío una imagen a color para ver sus características

estadisticas(img_GRAY)  # Envío una imagen Blanco y Negro para ver sus características
```
